```python
import random

def gerar_tuplas_aleatorias(num_tuplas=24, tamanho_tupla=7, minimo=1, maximo=31):
  """Gera uma lista de tuplas com números aleatórios."""
  tuplas = []
  for _ in range(num_tuplas):
    tupla = tuple(random.randint(minimo, maximo) for _ in range(tamanho_tupla))
    tuplas.append(tupla)
  return tuplas

def calcular_probabilidades(tuplas):
  """Calcula a probabilidade de ocorrência de cada número."""
  ocorrencias = {}
  total_numeros = 0
  for tupla in tuplas:
    for numero in tupla:
      ocorrencias[numero] = ocorrencias.get(numero, 0) + 1
      total_numeros += 1

  probabilidades = {}
  for numero, ocorrencia in ocorrencias.items():
    probabilidades[numero] = ocorrencia / total_numeros
  return probabilidades

def gerar_tuplas_probabilisticas(probabilidades, num_tuplas=12, tamanho_tupla=7):
  """Gera tuplas com base nas probabilidades calculadas."""
  numeros = list(probabilidades.keys())
  pesos = list(probabilidades.values())
  tuplas_probabilisticas = []
  for _ in range(num_tuplas):
    tupla = tuple(random.choices(numeros, weights=pesos, k=tamanho_tupla))
    tuplas_probabilisticas.append(tupla)
  return tuplas_probabilisticas

# Gerar 24 tuplas aleatórias
tuplas_aleatorias = gerar_tuplas_aleatorias()

# Calcular probabilidades
probabilidades = calcular_probabilidades(tuplas_aleatorias)

# Gerar 12 tuplas probabilísticas
tuplas_probabilisticas = gerar_tuplas_probabilisticas(probabilidades)

# Imprimir resultados
print("Tuplas Aleatórias:")
for tupla in tuplas_aleatorias:
  print(tupla)

print("\nProbabilidades:")
for numero, probabilidade in probabilidades.items():
  print(f"{numero}: {probabilidade:.4f}")

print("\nTuplas Probabilísticas:")
for tupla in tuplas_probabilisticas:
  print(tupla)
```
